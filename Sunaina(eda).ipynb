```python
dropping.info() #seeing if all the variables have same data type so that one can easily predict using ML models
```

```python
dropping["spends"] = dropping.spends.astype(int)
dropping["impressions"] = dropping.impressions.astype(int)
dropping["clicks"] = dropping.clicks.astype(int)

dropping.info()

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
ab=pd.get_dummies(dropping['campaign_platform'])
bc=pd.get_dummies(dropping['campaign_type'])
cd=pd.get_dummies(dropping['communication_medium'])
de = pd.get_dummies(dropping['subchannel'])
df2 = pd.concat((dropping,ab),axis=1)
df3 = pd.concat((df2,bc),axis=1)
df4 = pd.concat((df3,cd),axis=1)
dropping = pd.concat((df4,de),axis=1)

dropping.head()
dropping.info()
```

```python
train = dropping[['Brand','Google Ads','Search','Search Keywords']]
```

```python
test = dropping['spends']
```

```python
X_train,X_test,y_train,y_test = train_test_split(train,test,test_size=0.3,random_state=2)
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_train,y_train)
```

```python
pred = regr.predict(X_test)
```
